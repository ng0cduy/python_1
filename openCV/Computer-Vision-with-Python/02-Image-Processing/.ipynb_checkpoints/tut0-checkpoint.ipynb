```python
%config Completer.use_jedi = False
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("../DATA/00-puppy.jpg")
# in OPENCV, read in B-G-R order
# in MATPLOTLIB, read in R-G-B order
```

```python
type (img)
```

```python
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
# read the image and then convert to grayscale
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray,cmap='gray')
plt.imshow(img_gray,cmap='magma')
```

```python
fix_gray = cv2.cvtColor(img_gray,cv2.COLOR_BGR2RGB)
plt.imshow(fix_gray)
```

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
resized_img = cv2.resize(fix_img,[1000,650]);
plt.imshow(resized_img)
```

```python
w_ratio = 0.8
h_ratio = 0.5
resized_img_scale = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio);
```

```python
plt.imshow(resized_img_scale)
```

## FLIP

```python
new_img = cv2.flip(fix_img,-1)
plt.imshow(new_img)
# 0 for horizontal flip, 1 for horizontal flip, -1 for both
```

```python
cv2.imwrite('new_dog.jpg',fix_img)
```
