```python
%config Completer.use_jedi = False
import numpy
import cv2
import PIL
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
img = cv2.imread("../DATA/dog_backpack.jpg")
plt.imshow (img)
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
plt.imshow(cv2.flip(fix_img,0))
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
rec1 = cv2.rectangle(fix_img,pt1=(200,400),pt2=(550,700),color=(255,0,0),thickness=10)
plt.imshow(rec1)
# cv2.imshow("rec1",rec1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
